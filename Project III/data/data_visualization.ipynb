```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('test.csv', header = 0, sep = ',')
```

```python
vals = df['Sentiment']
```

```python
dic = {'positive': 0, 'negative': 0}
```

```python
for i in vals:
    if i == 1:
        dic['positive'] += 1
    else:
        dic['negative'] += 1
```

```python
key = dic.keys()
val = dic.values()
```

```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}% ({v:d})'.format(p=pct,v=val)
    return my_autopct
```

```python
plt.pie(val, labels = key, autopct=make_autopct(val))
plt.show()
```
