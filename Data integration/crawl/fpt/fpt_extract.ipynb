```python
import csv
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
browser = webdriver.ChromeOptions()
prefs = {"profile.default_content_setting_values.notifications" : 2} # Tắt thông báo Chrome
browser.add_experimental_option("prefs",prefs)
browser = webdriver.Chrome(chrome_options=browser, executable_path=r'D:\Tai Lieu\HUST-Study\20202\Tích hợp dl\nguyen\chromedriver.exe')
```

```python
df = pd.read_csv('test/3.csv')
links = df['link'].values.tolist()
```

```python
len(links)
```

```python
data = {'Tên': [], 'Link': [], 'Giá': [], 'Ảnh': [], 'Màu': [], 'Màn hình': [], 'Camera sau': [], 'Camera Selfie': [], 'RAM ': [], 'Bộ nhớ trong': [], 'CPU': [], 'GPU': [], 'Dung lượng pin': [], 'Thẻ sim': [], 'Hệ điều hành': [], 'Xuất xứ': [], 'Thời gian ra mắt': []}

detail = list(data.keys())
detail = detail[5:17]
```

```python
def extract(link):
    browser.get(link)

    # Link
    data['Link'].append(link)
    # print(link)

    #Name
    try:
        name = browser.find_element_by_class_name('st-name').text
        no = browser.find_element_by_class_name('st-sku').text
    except:
        name = browser.find_element_by_class_name('fs-dttname').text
        no = browser.find_element_by_class_name('nosku').text
    
    name = name.replace(no, '')
    data['Tên'].append(name)
    # print(name)

    #Price
    try:
        price = browser.find_element_by_class_name('st-price-main').text.replace('.', '').replace('đ', '').replace('-', '')
    except:
        price = ''

    data['Giá'].append(price)
    # print(price)

    #Color
    try:
        colors = browser.find_elements_by_xpath('//div[@class="st-select-color"]/*')
        if colors:
            color = []
            for i in range(len(colors)):
                c = colors[i].text.replace('-', '')
                if c != "":
                    color.append(c)
            color= ', '.join(str(i) for i in color)
            data['Màu'].append(color)
        else:
            raise Exception
    except:
        color = ''
        data['Màu'].append(color)
    
    # Image
    img = browser.find_element_by_xpath('//div[@class="swiper-wrapper js--slide--full"]/div/img').get_attribute('src')
    data['Ảnh'].append(img)

    #Detail
    try:
        key = browser.find_elements_by_xpath('//table[@class="st-pd-table"]/tbody/*/td[position()=1]')
        value = browser.find_elements_by_xpath('//table[@class="st-pd-table"]/tbody/*/td[position()=2]')
        if key:
            att = []
            for i in range(len(key)):
                k = key[i].text
                att.append(k)
                data[k].append(value[i].text.replace('-', ''))
            for i in range(len(detail)):
                if detail[i] not in att:
                    data[detail[i]].append('')
        else:
            raise Exception
    except:
        for i in range(len(detail)):
            data[detail[i]].append('')
    
    return data
```

```python
for i in range(len(links)):
    print(i)
    data = extract(links[i])
```

```python
browser.quit()
```

```python
df = pd.DataFrame.from_dict(data)
df
```

```python
df.to_csv('test/o3.csv', index = False)
```
