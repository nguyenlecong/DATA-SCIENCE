```python
import csv
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
browser = webdriver.ChromeOptions()
prefs = {"profile.default_content_setting_values.notifications" : 2} # Tắt thông báo Chrome
browser.add_experimental_option("prefs",prefs)
browser = webdriver.Chrome(chrome_options=browser, executable_path=r'D:\Tai Lieu\HUST-Study\20202\Tích hợp dl\nguyen\chromedriver.exe')
```

```python
browser.get('https://www.thegioididong.com/dtdd#i:6')
```

```python
items = browser.find_elements_by_xpath('//ul[@class="homeproduct item2020 "]/*/a')

links = []
i=0
for item in items:
    i+=1
    link = item.get_attribute('href')
    links.append(link)

remove = ['https://www.thegioididong.com/dtdd/vivo-x60-pro', 'https://www.thegioididong.com/dtdd/oppo-reno5-marvel', 'https://www.thegioididong.com/dtdd/vivo-y72-5g', 'https://www.thegioididong.com/dtdd/realme-8', 'https://www.thegioididong.com/dtdd/realme-8-pro', 'https://www.thegioididong.com/dtdd/xiaomi-mi-11-lite-4g', 'https://www.thegioididong.com/dtdd/oppo-a74']
links = [i for i in links if i not in remove]
```

```python
len(links)
```

```python
data = {'Tên': [], 'Link': [], 'Giá': [], 'Ảnh': [], 'Màu': [], 'Màn hình': [], 'Camera sau': [], 'Camera trước': [], 'Chip': [], 'RAM': [], 'Bộ nhớ trong': [], 'SIM': [], 'Pin': []}

detail = list(data.keys())
detail = detail[5:15]
```

```python
def extract(link):
    browser.get(link)

    # Link
    data['Link'].append(link)

    # Name
    name = browser.find_element_by_xpath('//div[@class="rowtop"]/h1').text
    data['Tên'].append(name)

    # Color
    color = []
    try:
        colors = browser.find_elements_by_xpath('//ul[@class="owl-carousel owl-theme tabscolor"]/div/div/*/li')
        if not colors:
            raise Exception
    except:
        colors = browser.find_elements_by_xpath('//ul[@class="owl-carousel owl-theme tabscolor"]/*')
    for c in colors:
        if c.get_attribute('data-order') != None:
            if c.text != '':
                color.append(c.text)
    color = ', '.join(str(i) for i in color)
    data['Màu'].append(color)

    # Price
    xpaths = ['//a[@class="tab-default active"]/div/strong', '//div[@class="area_price "]/strong', '//div[@class="area_price notapply"]/strong']
    for i in range(len(xpaths)):
        try:
            price = browser.find_element_by_xpath(xpaths[i])
        except:
            pass

    price = price.text.replace('.', '').replace('₫', '').replace('*', '').replace(' ', '')
    data['Giá'].append(price)

    # Image:
    img = browser.find_element_by_xpath('//div[@class="icon-position"]/img').get_attribute('src')
    data['Ảnh'].append(img)

    # Detail
    att = []
    key = browser.find_elements_by_xpath('//div[@class="tableparameter"]/ul/*/span')
    value = browser.find_elements_by_xpath('//div[@class="tableparameter"]/ul/*/div')
    for i in range(len(key)):
        k = key[i].text.replace(':', '')
        att.append(k)
        if k in detail:
            data[k].append(value[i].text)
    for i in range(len(detail)):
        if detail[i] not in att:
            data[detail[i]].append('')

    return data
```

```python
for i in range(len(links)):
    print(i)
    data = extract(links[i])
```

```python
browser.quit()
```

```python
df = pd.DataFrame.from_dict(data)
df
```
