```python
import csv
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
browser = webdriver.ChromeOptions()
prefs = {"profile.default_content_setting_values.notifications" : 2} # Tắt thông báo Chrome
browser.add_experimental_option("prefs",prefs)
browser = webdriver.Chrome(chrome_options=browser, executable_path=r'D:\Tai Lieu\HUST-Study\20202\Tích hợp dl\nguyen\chromedriver.exe')
```

```python
browser.get('https://clickbuy.com.vn/danh-muc/dien-thoai/')
```

```python
more = True
while more:
    try:
        moreproduct = browser.find_element_by_xpath('//div[@id="sb-infinite-scroll-load-more-1"]/a[text()="Xem thêm"]')
        moreproduct.click()
    except:
        more = False
```

```python
links = []
item = browser.find_elements_by_xpath('//li[@class="style-flex-colum col-6 col-sm-4 col-lg-4 ex_p2"]/a')
for item in item:
    link = item.get_attribute('href')
    links.append(link)
remove = ['https://clickbuy.com.vn/xiaomi-mi6x-ram-4gb-32gb-mi-a2/']
links = [i for i in links if i not in remove]
```

```python
len(links) # 157
```

```python
data = {'Tên': [], 'Link': [], 'Giá': [], 'Ảnh': [], 'Màu': [], 'Bộ nhớ trong': [], 'RAM': [], 'CPU': [], 'Kích thước màn hình': [], 'Độ phân giải màn hình': [], 'Camera chính': [], 'Camera phụ': [], 'Hệ điều hành': [], 'Thẻ sim': [], 'Dung lượng pin': []}
detail = list(data.keys())
detail = detail[5:17]
```

```python
def extract(link):
    browser.get(link)

# Link
    data['Link'].append(link)

# Name
    name = browser.find_element_by_xpath('//div[@class="ex_sgtitle"]/h1').text
    data['Tên'].append(name)

# Price
    try:
        try:
            price = browser.find_element_by_xpath('//p[@class="price"]/ins/span').text.replace('₫', '').replace(' ', '').replace(',', '')
        except:
            price = browser.find_element_by_xpath('//p[@class="price"]/span').text.replace('₫', '').replace(' ', '').replace(',', '')
    except:
        price = ''
    data['Giá'].append(price)

# Color
    color = []
    try:
        colors = browser.find_elements_by_xpath('//div[@class="tawcvs-swatches"]/*')
        for c in colors:
            c = c.text.split("\n")[0]
            color.append(c)
        color = ', '.join(str(i) for i in color)
    except:
        color = ''
    data['Màu'].append(color)

# Image
    img = browser.find_element_by_xpath('//figure[@class="woocommerce-product-gallery__wrapper"]/div/a').get_attribute('href')
    data['Ảnh'].append(img)

# Detail
    att = []
    key = browser.find_elements_by_xpath('//table[@class="woocommerce-product-attributes shop_attributes"]/tbody/*/th')
    value = browser.find_elements_by_xpath('//table[@class="woocommerce-product-attributes shop_attributes"]/tbody/*/td')
    for i in range(len(key)):
        k = key[i].text
        att.append(k)
        if k in detail:
            v = value[i].text.replace('\n', '')
            data[k].append(v)
    for i in range(len(detail)):
        if detail[i] not in att:
            data[detail[i]].append('')

    return data
```

```python
for i in range(len(links)):
    data = extract(links[i])
    print(i)
```

```python
browser.quit()
```

```python
df = pd.DataFrame.from_dict(data)
df
```
